```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/synthetic_profile_data.csv')
print(df.head())
```

```python
# Create a quick plot to visualize it
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
ax1.plot(df['distance_km'], df['topography_m'])
ax1.set_ylabel('Topography [m]')
ax1.grid(True)
ax2.plot(df['distance_km'], df['bouguer_mgal'])
ax2.set_ylabel('Bouguer Anomaly [mGal]')
ax2.set_xlabel('Distance [km]')
ax2.grid(True)
plt.show()
```
